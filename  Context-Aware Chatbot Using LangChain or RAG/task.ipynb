```python
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings # Note the change here
from langchain_community.vectorstores import Chroma

# 1. Setup Embeddings
# This uses the community version which has broader version compatibility
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# 2. Load your CV
CV_NAME = "Usman_CV.pdf"

if os.path.exists(CV_NAME):
    loader = PyPDFLoader(CV_NAME)
    data = loader.load()
    
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    chunks = text_splitter.split_documents(data)

    # 3. Create Local Database
    vector_db = Chroma.from_documents(
        documents=chunks, 
        embedding=embeddings,
        persist_directory="./local_db"
    )
    print("✅ Success! 'local_db' created.")
else:
    print(f"❌ Error: '{CV_NAME}' not found.")
```
