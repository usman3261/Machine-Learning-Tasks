```python
import pandas as pd
import torch
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer

# 1. Load Data
df = pd.read_csv("ag_news.csv")
dataset = Dataset.from_pandas(df)

# 2. Setup DistilBERT
model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

def tokenize_func(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_func, batched=True)

# 3. Setup Model
model = AutoModelForSequenceClassification.from_pretrained(model_ckpt, num_labels=4)

# 4. Training Arguments (Stripped down)
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=4,
    save_strategy="no",        # Don't save checkpoints to save space
    report_to="none"           # No external logging
)

# 5. Initialize Trainer (No compute_metrics = No NumPy needed)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
)

# 6. Run Training
print("Training started... this may take a few minutes.")
trainer.train()

# 7. Save Model and Tokenizer
model.save_pretrained("./news_model")
tokenizer.save_pretrained("./news_model")
print("✅ Success! Model saved to ./news_model folder.")
```
